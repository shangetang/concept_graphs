```python
import numpy as np
import matplotlib.pyplot as plt

file_path = '/scratch/gpfs/st3812/projects/concept_graphs/output/single-body_2d_3classes/H32-train1/5000_1.6_256_500_100_0.0001_None_1500_2.0_1/image_000_ep95.npz'
# file_path = '/scratch/gpfs/st3812/projects/concept_graphs/output/single-body_2d_3classes/H32-train1/5000_1.6_256_500_100_0.0001_None_1500_2.0_1/gen_store_000_ep99.npz'

# Load the .npz file
data = np.load(file_path)

```

```python
# List all arrays in the file
keys = data.files
print("Keys in NPZ file:", keys)

# Plot a few images from each array
for key in keys:
    images = data[key]  # Shape: (25, 64, 3, 28, 28)
    print(f"Shape of {key}:", images.shape)

    # Loop through the first batch and first few images
    for i in range(5):  # Display 5 images
        # img = images[0, i]  # First batch, ith image (Shape: 3, 28, 28)
        img = images[i] # ith image (Shape: 3, 28, 28)
        print(img.max())
        print(img.min())

        # Transpose to (28, 28, 3) for imshow
        img = np.transpose(img, (1, 2, 0))
        # img = (img +1) /2
        # img = img.clip(0, 1)
        
        plt.figure()
        plt.title(f"{key} - Image {i+1}")
        plt.imshow(img)
        plt.axis('off')

plt.show()

# Close the file
data.close()
```
